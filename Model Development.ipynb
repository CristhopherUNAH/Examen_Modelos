```python
# Sacar la mediana y llenar los datos nulos de residual sugar
aux = wine['residual sugar'].median()
wine['residual sugar'].fillna(aux, inplace=True)
```

```python
columnas = [
    "free sulfur dioxide"        
    , "total sulfur dioxide"
    , "density"
    , "alcohol",
    'residual sugar'
]

# Entrenamiento 
x = wine.loc[ : , columnas ].values
x = StandardScaler().fit_transform(x)
np.nan_to_num(x)
x = np.around( x , 0 )
```

```python
pca = PCA( n_components=2)
c = pca.fit_transform(x)
pcadf = pd.DataFrame(data=c ,columns=['c1','c2'])
plt.scatter( pcadf['c1'] , pcadf['c2'] ) 

# la similitud es muy constante vale mucho la pena para hacer uno para while y red
```

```python
#3
# Hacer la revicion de cuantos Cluster se necesitan segun la grafica
sse = {}
for k in range(1, 10):
    kmeans = KMeans(n_clusters=k, max_iter=1000).fit(pcadf)    
    sse[k] = kmeans.inertia_
plt.figure()
plt.plot(list(sse.keys()), list(sse.values()))
plt.xlabel("Number of cluster")
plt.ylabel("SSE")
plt.show()
```

```python
km = KMeans( n_clusters=3  ).fit(pcadf)
km.cluster_centers_
cetroids = km.cluster_centers_
km.labels_
plt.scatter( pcadf['c1'] , pcadf['c2'] , c=km.labels_.astype(float) , s=50, alpha=0.5 ) 
```

```python
# Tiene algo en comun los 3 clusteres visto en el grafica
wine2['cluster'] = km.labels_
wine2
```

```python
#4, #5 y #6
# eliminacion de lo pedido por el enunciado y de type por la limpiaza previa que se tiene que hacer
# agregar la nueva columna y agregar sus datos 
wine3 = wine.drop(columns = ["free sulfur dioxide","total sulfur dioxide","density","alcohol","residual sugar"])
wine3['Nueva_variable'] = np.where(wine3['quality'] >= 5,1,0)
wine3 = wine3.drop(columns = ["quality", 'type'])
wine_nuevo = wine3
wine_nuevo
```

```python
# 7 
# A)
# llenar los null con la mediana por los errores que pueden dar en el futuro por los nulos 
# Regresion Logistica
aux1 = wine_nuevo['fixed acidity'].median()
wine_nuevo['fixed acidity'].fillna(aux, inplace=True)
aux2 = wine_nuevo['volatile acidity'].median()
wine_nuevo['volatile acidity'].fillna(aux, inplace=True)
aux3 = wine_nuevo['citric acid'].median()
wine_nuevo['citric acid'].fillna(aux, inplace=True)
aux4 = wine_nuevo['chlorides'].median()
wine_nuevo['chlorides'].fillna(aux, inplace=True)
aux5 = wine_nuevo['pH'].median()
wine_nuevo['pH'].fillna(aux, inplace=True)
aux6 = wine_nuevo['sulphates'].median()
wine_nuevo['sulphates'].fillna(aux, inplace=True)

features_col = wine_nuevo.columns.values[0:6]
x = wine_nuevo[features_col].values
y = wine_nuevo["Nueva_variable"].values
X_train, X_test, Y_train, Y_test = train_test_split(x,y,test_size = 0.25, random_state=0)

log_model = LogisticRegression(solver='lbfgs', max_iter=1000)
log_model.fit(X_train,Y_train)
```

```python
y_pred  = log_model.predict(X_test)
y_pred
```

```python
print(metrics.accuracy_score(Y_test,y_pred))
print(precision_score(Y_test, y_pred))
print(recall_score(Y_test, y_pred))
print(f1_score(Y_test, y_pred))
```

```python
# B)
# NN
x_train = X_train 
x_test = X_test
y_train = Y_train
y_test =  Y_test
NN = MLPClassifier(random_state=1, max_iter=3000)
NN.fit(x_train, y_train)
```

```python
y_pred = NN.predict(x_test)
accuracy = accuracy_score(y_test,y_pred)*100
confusion_mat = confusion_matrix(y_test,y_pred)
print(metrics.accuracy_score(y_test,y_pred))
print(precision_score(y_test, y_pred))
print(recall_score(y_test, y_pred))
print(f1_score(y_test, y_pred))
```

```python
# C)
# Aroboles de desiciones
tree_one = tree.DecisionTreeClassifier(random_state=1)
tree_one = tree_one.fit( x_train , y_train )
tree_one.tree_.max_depth
```

```python
y_pred = tree_one.predict(x_test)
print(metrics.accuracy_score(y_test,y_pred))
print(precision_score(y_test, y_pred))
print(recall_score(y_test, y_pred))
print(f1_score(y_test, y_pred))
```

```python
# D)
# Random Forest => ganador 
rf = RandomForestRegressor(n_estimators = 1000, random_state = 42)
rf.fit(x_train , y_train);
y_pred = rf.predict(x_test) >= 0.5
```

```python
print(metrics.accuracy_score(y_test,y_pred))
print(precision_score(y_test, y_pred))
print(recall_score(y_test, y_pred))
print(f1_score(y_test, y_pred))
```
